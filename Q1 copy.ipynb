```python
# Import relevant packages
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
```

```python
# Create the BN model which contains edges of the DAG graph
# Note that CPDs are not required; D-separation depends only on the graph structure

## write your code here [2 points]
model = DiscreteBayesianNetwork([
    ('A','B'), ('A','C'), ('B','D'), ('B','C'),
    ('G','D'), ('G','H'), ('D','E'), ('H','E'),
    ('E','F'), ('F','C'), ('F','I'), ('H','I')
])

```

```python
# add d-separation method
from types import MethodType
def _is_dseparated(self, X, Y, observed=None):
    # allow passing single node names like 'X1'
    if isinstance(X, str):
        X = [X]
    if isinstance(Y, str):
        Y = [Y]
    observed = observed or []
    return not any(
        self.is_dconnected(x, y, observed=observed)
        for x in X
        for y in Y
    )

model.is_dseparated = MethodType(_is_dseparated, model)
```

```python
# Use the d-separation method to check the conditional independence statements in (1a) - (1d)

## write your code here [2 points]
# Check (1a) whether B and G are d-separated
print(f"Are B and G d-separated?: {model.is_dseparated('B', 'G', [])}") 


# Check (1b) whether C and H are d-separated given I
print(f"Are C and H d-separated given [I]?: {model.is_dseparated('C', 'H', ['I'])}")


# Check (1c) whether A and H are d-separated given {B, F}
print(f"Are A and H d-separated given [B, F]?: {model.is_dseparated('A', 'H', ['B','F'])}")


# Check (1d) whether A and {G, I} are d-separated given {B, C}
print(f"Are A and [G, I] d-separated given [B, C]?: {model.is_dseparated('A', ['G','I'], ['B','C'])}")
```
